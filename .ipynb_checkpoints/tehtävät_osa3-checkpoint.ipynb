# Tehtävä 1.
## Aihe: Uutisotsikon klassifiointi - Clickbait vai ei?
2,5 pistettä

Datasetti:
https://github.com/bhargaviparanjape/clickbait/tree/master/dataset

1. Lataa clickbait ja ei-clickbait otsikot clickbait_data.txt ja non_clickbait_data.txt tiedostoista data - kansiosta
2. Leimaa otsikot 0 tai 1 luokkaan (clickbait vai ei)
2. Jaa data koulutus- ja testidataan (80% koulutusdataa 20% testidataa jako)
4. Luo RNN - malli, joka ennustaa, onko otsikko clickbait vai ei
5. Tulosta mallin tarkkuus evaluate - funktiolla
6. Aja "Tehtävän vastaukset" solu

Vinkkejä:
* Käytä materiaaleista tuttua Tokenizer - luokkaa
* Luo otsikoista samanpituiset sekvenssit

```python
# Tehtävän toteutus
import pandas as pd
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt

clickbait_file = "data/clickbait_data.txt"
non_clickbait_file = "data/non_clickbait_data.txt"

df1 = pd.read_csv(clickbait_file, names=['Title'], delimiter = "\n") #15999
df1['isclickbait'] = 0
#df1['Title'] = list(map(lambda x: tf.keras.preprocessing.text.one_hot(x, 100)))

df2 = pd.read_csv(non_clickbait_file, names=['Title'], delimiter = "\n") #16001
df2['isclickbait'] = 1
      
train_0, train_1 = df1.sample(frac=0.8, random_state=1), df2.sample(frac=0.8, random_state=1)
#print(len(train_0)) #12799
#print(len(train_1)) #12799

test_0, test_1 = df1.drop(train_0.index), df2.drop(train_1.index)
#print(len(test_0)) #3200
#print(len(test_1)) #3202

frames1 = [train_0, train_1]
df_train = pd.concat(frames1)
df_train = df_train.sample(frac=1).reset_index(drop=True)
#print(df_train)

frames2 = [test_0, test_1]
df_test = pd.concat(frames2)
df_test = df_test.sample(frac=1).reset_index(drop=True)
#print(df_test)

#X's: 12799 + 3200 = 15999
#y's: 12799 + 3202 = 16001 => both match!

train_y, test_y = df_train['isclickbait'], df_test['isclickbait']

num_words = 1000

tokenizer = tf.keras.preprocessing.text.Tokenizer(num_words=num_words)
tokenizer.fit_on_texts(df_train['Title'])
#print(df['Title'])

train_seq = tokenizer.texts_to_sequences(df_train['Title'])
#print(len(train_seq)) #12799
test_seq = tokenizer.texts_to_sequences(df_test['Title'])
#print(len(test_seq)) #3200

train_padded = tf.keras.preprocessing.sequence.pad_sequences(train_seq, maxlen=10, padding='post')
test_padded = tf.keras.preprocessing.sequence.pad_sequences(test_seq, maxlen=10, padding='post')

#print(train_padded.shape)

input_layer = tf.keras.Input(shape=(10, ))
embedding_layer = tf.keras.layers.Embedding(1000, 2, input_length=10)(input_layer)
lstm = tf.keras.layers.LSTM(10, return_sequences=True)(embedding_layer)
lstm2 = tf.keras.layers.LSTM(8, return_sequences=True)(lstm)
lstm3 = tf.keras.layers.LSTM(6)(lstm2)
dense = tf.keras.layers.Dense(2, activation='sigmoid')(lstm3)
dense2 = tf.keras.layers.Dense(2, activation='softmax')(dense)

lstm_model = tf.keras.Model(inputs=input_layer,
                          outputs=dense2)

lstm_model.compile(loss='sparse_categorical_crossentropy',
                     optimizer=tf.keras.optimizers.Adam(learning_rate=0.001),
                     metrics=['accuracy'])
lstm_model.summary()

print(train_padded.shape)
print(train_y.shape)

lstm_model.fit(train_padded, train_y, epochs=10, batch_size=100, validation_data=(test_padded, test_y))

results = lstm_model.evaluate(test_padded, test_y)
```

```python
# Tehtävän vastaukset. Huom! Älä muokkaa tätä solua, vaan aja se, kun olet suorittanut tehtävän. Sijoita results - muuttujaan funktion model.evaluate() tulos.
# Muista määrittää model.compile() - funktioon seurattavaksi suureeksi metrics=['accuracy'], jotta näät, kuinka suuri osa neuroverkon ennustuksista on oikein.
print(f"Test Loss:{results[0]} Test Accuracy:{results[1]*100}%")
```

# Tehtävä 2.
### Aihe: Aikasarjaennustaminen - Lämpötila ja sähkönkulutus
2,5 pistettä

Data - kansiossa tiedostoissa 'lampotila_2010.csv' ja 'sahkonkulutus_2010.csv' on vuoden 2010 tammikuun lämpötila Jyväskylän lentoasemalta ja koko Suomen sähkönkulutus. Tavoitteena on luoda neuroverkkomalli, joka ennustaa sähkönkulutuksen ja lämpötilan avulla tulevaa sähkönkulutusta.

1. Ennusta 24 tunnin historiadatalla 24 tuntia eteenpäin
2. Kouluta neuroverkko vuoden 2010 tammikuun datalla.
3. Luo testidatasetti 'lampotila_2011.csv' ja 'sahkonkulutus_2011.csv' tiedostoista
4. Ennusta mallilla vuoden 2011 tammikuun sähkönkulutusta. Piirrä oikeat arvot ja ennustetut arvot viivakaavioon.
5. Aja "Tehtävän vastaukset" solu

Vinkkejä:
* Muuta aikasarja sekvenssimuotoon käyttäen multivariate_data - funktiota materiaaleista.
    * Mitkä ovat muuttujia, joiden avulla ennustetaan, ja mitkä ovat ennustettavia arvoja?
* Dataa on vähän, joten ennustustulokset saattavat olla huonoja
    * Tärkeintä on, että idea on oikein
    * Käytä Dropout - kerroksia ja vaihda LSTM - kerroksissa solujen määrää

```python
# Tehtävän toteutus
import pandas as pd
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler

temperature2010 = 'data/lampotila_2010.csv'
electricity2010 = 'data/sahkonkulutus_2010.csv'

electricity2011 = 'data/sahkonkulutus_2011.csv'
temperature2011 = 'data/lampotila_2010.csv'

#2010 datat
df3 = pd.read_csv(temperature2010, delimiter=',', parse_dates=[['Vuosi','Kk','Pv','Klo']]) #744 rows × 3 columns
df3 = df3.rename(columns={'Vuosi_Kk_Pv_Klo':'Date', 'Aikavyöhyke' : 'Timezone','Ilman lämpötila (degC)':'Temperature (degC)'})
df3 = df3.drop(['Timezone'], axis=1) #744 rows × 2 columns
#df3

df4 = pd.read_csv(electricity2010, delimiter=',') #744 rows × 5 columns
df4 = df4.rename(columns={'Alkuaika UTC':'Start time UTC',
                          'Lopetusaika UTC' : 'End time UTC',
                          'Alkuaika UTC+02:00':'Start time UTC+02:00',
                         'Lopetusaika UTC+02:00':'End time UTC+2:00',
                         'Sähkön kulutus Suomessa':'Electricity consumption in Finland (Hz)'})
#df4 #744 rows × 5 columns

frames1 = [df3, df4]
df5 = pd.concat(frames1, axis=1, join="inner")
#df5 = pd.concat(frames)
#df5 #744 rows × 8 columns
df5 = df5.fillna(method='bfill')
df5 = df5.set_index('Date')

#2011 datat
df6 = pd.read_csv(temperature2011, delimiter=',', parse_dates=[['Vuosi','Kk','Pv','Klo']])
df6 = df6.rename(columns={'Vuosi_Kk_Pv_Klo':'Date', 'Aikavyöhyke' : 'Timezone','Ilman lämpötila (degC)':'Temperature (degC)'})
df6 = df6.drop(['Timezone'], axis=1) #744 rows × 2 columns
#df6

df7 = pd.read_csv(electricity2011, delimiter=',') #744 rows × 5 columns
df7 = df7.rename(columns={'Alkuaika UTC':'Start time UTC',
                          'Lopetusaika UTC' : 'End time UTC',
                          'Alkuaika UTC+02:00':'Start time UTC+02:00',
                         'Lopetusaika UTC+02:00':'End time UTC+2:00',
                         'Sähkön kulutus Suomessa':'Electricity consumption in Finland (Hz)'})
#df7 #744 rows × 5 columns

frames2 = [df6, df7]
df8 = pd.concat(frames2, axis=1, join="inner")
df8 = df8.fillna(method='bfill')
#df8 #744 rows × 7 columns

scaler = MinMaxScaler(feature_range=(0,1))
train_dataset = scaler.fit_transform(df5[['Temperature (degC)', 'Electricity consumption in Finland (Hz)']].values)
test_dataset = scaler.fit_transform(df8[['Temperature (degC)', 'Electricity consumption in Finland (Hz)']].values)
#print(train_dataset.shape) #(744, 2)
#print(test_dataset.shape) #(744, 2)

def multivariate_data(dataset, target, start_index, end_index, history_size, target_size, step, single_step=False):
    data = []
    labels = []
    start_index = start_index + history_size
    if end_index is None:
        end_index = len(dataset) - target_size

    for i in range(start_index, end_index):
        indices = range(i-history_size, i, step)
        data.append(dataset[indices])
        
        if single_step:
            labels.append(target[i+target_size])
        else:
            labels.append(target[i:i+target_size])
            
    return np.array(data), np.array(labels)
past_history = 24
future_target = 24
STEP = 1
train_X, train_y = multivariate_data(train_dataset[:, 0], train_dataset[:, 1], 0,
                                                 len(train_dataset), past_history,
                                                 future_target, STEP)
test_X, test_y = multivariate_data(test_dataset[:, 0], test_dataset[:, 1], 0,
                                             len(test_dataset), past_history,
                                             future_target, STEP)

train_X = train_X.reshape(len(train_X), 24, 1)
test_X = test_X.reshape(len(test_X), 24, 1)
print (f'Single window of past history : {train_X[0].shape}')
print (f'Target to predict : {train_y[0].shape}')

train_y = tf.keras.preprocessing.sequence.pad_sequences(train_y, maxlen=24, padding='post', dtype='float64')
test_y = tf.keras.preprocessing.sequence.pad_sequences(test_y, maxlen=24, padding='post', dtype='float64')

input_layer = tf.keras.Input(shape=(train_X.shape[1],train_X.shape[2]))
lstm = tf.keras.layers.LSTM(32, return_sequences=True)(input_layer)
lstm2 = tf.keras.layers.LSTM(26, return_sequences=True)(lstm)
lstm3 = tf.keras.layers.LSTM(18, return_sequences=True)(lstm2)
dropout3 = tf.keras.layers.Dropout(.2)(lstm3)
bilstm = tf.keras.layers.Bidirectional(tf.keras.layers.LSTM(20, return_sequences=True))(dropout3)
lstm4 = tf.keras.layers.LSTM(12, return_sequences=True)(bilstm)
lstm5 = tf.keras.layers.LSTM(6, activation='relu')(lstm4)
dense = tf.keras.layers.Dense(24)(lstm5)
dense2 =tf.keras.layers.Dense(24)(dense)
model = tf.keras.Model(inputs=input_layer,
                      outputs=dense2)

model.compile(optimizer=tf.keras.optimizers.Adam(lr=0.0001), loss='mse', metrics=['mse', 'mae'])
model.fit(train_X, train_y, validation_data=(test_X, test_y), epochs=1, batch_size=1)
```

```python
nums_preds = []
nums_real = []
index = 0
# Käydään toistorakenteessa läpi ennustukset
predictions = model.predict(test_X,verbose=0)
for x in range(len(predictions) // len(predictions[0])):
    for num in predictions[index]:
        nums_preds.append(num)
    for num in test_y[index]:
        nums_real.append(num)
    # Siirrytään 24 tuntia eteenpäin, kun sekvenssi oikeista arvoista ja ennustetuista arvoista on lisätty listaan
    index += len(predictions[index]) - 1
fig, ax = plt.subplots(figsize=(17,11))
ax.plot(nums_preds,label='Predictions')
ax.plot(nums_real,label='True')
ax.legend()
plt.show()
```

```python
# Tehtävän toteutus
import pandas as pd
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split

data = 'data/creditcard.csv'

df_data = pd.read_csv(data, delimiter=',')
df_data

X = df_data.iloc[:, :30]
y = df_data[['Class']]

train_X, test_X, train_y, test_y = train_test_split(X, y, test_size=0.3, random_state=1, shuffle=True, stratify=df_data[['Class']])

scaler = MinMaxScaler()
train_X = scaler.fit_transform(train_X)
test_X = scaler.transform(test_X)

test_X_normal = test_X[:5000]
test_X_modified = test_X[5000:]
test_X_modified = scaler.transform(test_X_modified)

features = train_X.shape[1]
input_layer = tf.keras.Input(shape=(features,))
encoder = tf.keras.layers.Dense(7,activation='relu')(input_layer)
decoder = tf.keras.layers.Dense(features, activation='sigmoid')(encoder) 
autoencoder = tf.keras.Model(inputs=input_layer,
                             outputs=decoder)

autoencoder.compile(optimizer=tf.keras.optimizers.Adam(lr=0.001),loss='mean_squared_error')
autoencoder.fit(train_X,train_X,epochs=2,batch_size=4)
```

```python
normal_predictions = autoencoder.predict(test_X)
mse_normal = np.mean((test_X - normal_predictions)**2,axis=1)
fig, ax = plt.subplots(figsize=(12,6))
ax.scatter(range(len(mse_normal)),mse_normal)
plt.show()
```
